```python
import csv
import re
from io import StringIO
from typing import Dict, List, Set, Optional, Tuple, Pattern
from attr import Factory, attrs, attrib
from pathlib import Path, PurePath
from more_itertools import split_before, strip, first, rstrip, groupby_transform, chunked, map_except, flatten
from typing import ClassVar
from itertools import groupby, starmap, tee, cycle

NoneStr = Optional[str]

dummy = Path('~/git/sample-sheet/tests/resources/paired-end-single-index-with-manifest.csv')

@attrs
class SampleSheet(object):
    header_pattern: ClassVar[Pattern] = re.compile('\[(.+)\]')
    required_sections: ClassVar[Set[str]] = set(['Header', 'Reads', 'Settings', 'Data'])
    
    @staticmethod
    def match_header(string: str) -> NoneStr:
        match = SampleSheet.header_pattern.match(string)
        return match.group(1) if match else None
    
    @staticmethod
    def _parse(text: str) -> None:
        lines: List[List[str]] = list(csv.reader(StringIO(text)))
        lines = [[i.strip() for i in row] for row in lines if ''.join(row).strip() is not '']
        grouped = groupby_transform(lines, lambda x: SampleSheet.match_header(first(x)), reducefunc=list)
        ordered = odicti(chunked([pair[i] for i, pair in zip(cycle([0, 1]), grouped)], 2))
        
        missing_sections = SampleSheet.required_sections - set(ordered)
        if len(missing_sections) != 0:
            raise KeyError(f'Required section(s) not found in sample sheet: ' + ', '.join(missing_sections))

        sample_header, *sample_records = ordered.pop('Data')
        #samples = [Sample(zip(sample_header, values)) for values in sample_records]
        
        #reads = [int(first(row)) for row in ordered.pop('Reads')]
        
        #print(SampleSheet(samples, reads=reads))
        return None

    @classmethod
    def read(cls, infile: Path) -> 'SampleSheet':
        cls._parse(infile.expanduser().absolute().read_text())
        return SampleSheet(samples=[])

SampleSheet.read(dummy)
```
